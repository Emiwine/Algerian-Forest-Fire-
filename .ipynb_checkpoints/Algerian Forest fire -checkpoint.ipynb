### ALGERIAN FOREST FIRE DATASET 



The dataset includes 244 instances that regroup a data of two regions of Algeria, namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attributes and 1 output attribute (class)
The 244 instances have been classified into fire (138 classes) and not fire (106 classes) classes

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset = pd.read_csv("Algerian_forest_fires_dataset_UPDATE.csv" , header=1 )
```

```python
dataset.head()
```

```python
dataset.info()
```

### DATA CLEANING 

```python
##missing values
dataset[dataset.isnull().any(axis = 1)]
```

The dataset is convert into two sets based on the region from 122th index , we can make a new column based on the region  

1 Beljaia Region dataset 

2 sidi-Bell abbes region dataset

```python
#here we create the new column because the data set is divided into two parts 
dataset.loc[:122 , "Region"] = 0 
dataset.loc[122: , "Region"] = 1
df = dataset
```

```python
df.head()
```

```python
df[['Region']] = df[["Region"]].astype(int) #here convert  the float datatype into int datatype 
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#Removing the null values 
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df =df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip() , #strip is basically remove the spaces in the columns 
df.columns
```

```python
df.info()
```

### Changes the required columns as integer datatype

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Changing the others columns to float datatype 

```python
objects = [features for features in df.columns if df[features].dtype ==  'O'] # here taking all the objects datatype 
```

```python
objects
```

```python
for i in objects:  #here converting all the object datatype into float except classes because it is categorical data type
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
objects
```

```python
df.head()
```

```python
# let save the cleaned dataset
df.to_csv("Algerian_forest_fires_cleaned_dataset.csv" , index =False)
```
